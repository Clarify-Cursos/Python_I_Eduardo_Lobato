```python
'''
 SISTEMA DE GERENCIAMENTO DE ESTOQUE
 NESTA ETAPA, VAMOS LIDAR APENAS COM O GERENCIAMENTO DO ESTOQUE DE PRODUTOS.

 ENTRADA:
 NOME DO PRODUTO
 QUANTIDADE DO PRODUTO
 PREÇO DO PRODUTO

 PROCESSAMENTO
 ADICIONAR UM NOVO PRODUTO AO ESTOQUE
 ATUALIZAR A QUANTIDADE DE UM PRODUTO NO ESTOQUE
 LISTAR TODOS OS PRODUTOS NO ESTOQUE COM SUAS QUANTIDADES E PREÇOS
 REMOVER UM PRODUTO DO ESTOQUE

 SAÍDA:
 LISTA DE PRODUTOS NO ESTOQUE COM SUAS RESPECTIVAS QUANTIDADES E PREÇOS
 '''

estoque = {}
menu = 0

def adicionar_produtos():
    nome = input('Digite o nome do produto: ')
    quantidade = int(input("digite a quantidade do produto: "))
    preco = float(input('Digite o preço do produto: '))
    estoque[nome] = {'nome :', nome, 'quantidade :', quantidade, 'preço :', preco}
    print(f'Produto {nome} adicionado com sucesso !')

def remover_produto():
    nome = input("Digite o nome do produto: ")
    if nome in estoque:
        del estoque[nome]
        print(f"Produto {nome} removido com sucesso")
    else:
        print("Produto não encontrado no estoque")

def listar_produtos():
    if not estoque:
        print('Estoque vazio')
    else:
        for nome, info in estoque.items():
            print(f"Produto:{info['nome']}, Quantidade{info['quantidade']}, Preço: R${info['preco']}")

def atualizar_produto():
    nome = input("Digite o nome do produto: ")
    if nome in estoque:
        operacao = int(input("Digite 1 para adicionar quantidade e 2 para remover quantidade:"))
        quantidade = int(input("Digite a quantidade:"))
        if operacao == 1:
            estoque[nome]["quantidade"] += quantidade
            print(f"Quantidade do produto {nome} aumentada com sucesso.")
        elif operacao == 2:
            if quantidade <= estoque[nome]["quantidade"]:
                estoque[nome]["quantidade"] -= quantidade
                print(f"Quantidade do produto {nome} diminuida com sucesso.")
            else:
                print("Operação não concluída por erro de estoque")
        else:
            print("Opção inválida. Insira 1 ou 2, apenas. " )
    else:
        print("Produto não encontrado no estoque")

while menu < 5:

    menu = int(input('''
                Selecione a operação:
                [1] ADICIONAR PRODUTO AO ESTOQUE
                [2] LISTAR TODOS OS PRODUTOS
                [3] ATUALIZAR A QUANTIDADE DE UM PRODUTO NO ESTOQUE
                [4] REMOVER UM PRODUTO
                [5] SAIR
                '''))

    match menu:
        case 1:
            adicionar_produtos()
        case 2:
            listar_produtos()
        case 3:
            atualizar_produto()
        case 4:
            remover_produto()
        case _:
            break
```

```python
# Dicionário para armazenar o estoque de produtos
estoque = {}

# Função para adicionar um novo produto ao estoque
def adicionar_produto(nome, quantidade, preco):
    if nome in estoque:
        print(f"O produto '{nome}' já existe no estoque. Atualizando a quantidade.")
        estoque[nome]['quantidade'] += quantidade
    else:
        estoque[nome] = {'quantidade': quantidade, 'preco': preco}
    print(f"Produto '{nome}' adicionado/atualizado com sucesso!")

# Função para atualizar a quantidade de um produto
def atualizar_quantidade(nome, quantidade):
    if nome in estoque:
        estoque[nome]['quantidade'] = quantidade
        print(f"Quantidade do produto '{nome}' atualizada com sucesso!")
    else:
        print(f"Produto '{nome}' não encontrado no estoque.")

# Função para listar todos os produtos no estoque
def listar_produtos():
    if estoque:
        print("\nLista de Produtos no Estoque:")
        for nome, dados in estoque.items():
            print(f"Produto: {nome} | Quantidade: {dados['quantidade']} | Preço: R$ {dados['preco']:.2f}")
    else:
        print("Não há produtos no estoque.")

# Função para remover um produto do estoque
def remover_produto(nome):
    if nome in estoque:
        del estoque[nome]
        print(f"Produto '{nome}' removido do estoque com sucesso!")
    else:
        print(f"Produto '{nome}' não encontrado no estoque.")

# Função principal que oferece o menu para o usuário
def menu():
    while True:
        print("\n--- Sistema de Gerenciamento de Estoque ---")
        print("1. Adicionar/Atualizar Produto")
        print("2. Atualizar Quantidade de Produto")
        print("3. Listar Todos os Produtos")
        print("4. Remover Produto")
        print("5. Sair")
        
        opcao = input("Escolha uma opção (1-5): ").strip()

        if opcao == '1':
            nome = input("Digite o nome do produto: ").strip()
            quantidade = int(input("Digite a quantidade do produto: "))
            preco = float(input("Digite o preço do produto: R$ "))
            adicionar_produto(nome, quantidade, preco)
        elif opcao == '2':
            nome = input("Digite o nome do produto: ").strip()
            quantidade = int(input("Digite a nova quantidade do produto: "))
            atualizar_quantidade(nome, quantidade)
        elif opcao == '3':
            listar_produtos()
        elif opcao == '4':
            nome = input("Digite o nome do produto que deseja remover: ").strip()
            remover_produto(nome)
        elif opcao == '5':
            print("Saindo do sistema...")
            break
        else:
            print("Opção inválida! Por favor, escolha uma opção entre 1 e 5.")

# Chama o menu principal
menu()
```
